## 將所有變數進行標準化

```python
normalized_share_MT = (share_MT_clustering - share_MT_clustering.mean()) / share_MT_clustering.std()
normalized_share_MT.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score as silhouette_score
```

## 依據惻隱係數判斷分群個數

```python
# investigate alternative numbers of clusters using silhouette score
silhouette_value = []
k = range(2,21)  # look at solutions between 2 and 20 clusters
for i in k:
    clustering_method = KMeans(n_clusters = i, random_state = 9999)
    clustering_method.fit(normalized_share_MT)
    labels = clustering_method.predict(normalized_share_MT)
    silhouette_average = silhouette_score(normalized_share_MT, labels)
    print('Silhouette Score for %i Clusters: %0.4f' % (i, silhouette_average))
```

## 選擇分群數4

```python
clustering_method = KMeans(n_clusters = 4, random_state = 9999)
clustering_method.fit(normalized_share_MT)
labels = clustering_method.predict(normalized_share_MT)
```

```python
share_MT['clusters']=labels
normalized_share_MT['clusters'] = labels
normalized_share_MT.head()
```

## 結論:
第一群:女生男生比例沒差很多，男生多於女生，年齡層31-35，大學或研究所，軍公教為主，每月可支配金額15000以上，大部分持有機車，有汽車，有機車駕照，沒使用過服務的比較多，只使用機車跟小客車，認知四項都有(全部值都沒有0)，使用目的每個比例都差不多

第二群:跟第三群條件差不多，但是認知部分(四個均為0)，使用過共享機車服務的人是最少的

第三群:女生男生比例沒差很多(女生比較多)，年齡層18-25，台中市居多，主要為大學(專)生，每月可支配金額5000到10000，大部分持有機車，但沒有汽車，有些沒有機車駕照，最常使用機車跟大眾交通運輸工具，認知四項都有(價格跟新服務的值都沒有0)，使用過共享機車服務的人最多，使用目的為通勤>觀光>娛樂

第四群:女生比較多，年齡層51以上，台中市為主，教育層度高中職以下的最多，職業多為服務業或退休或家管，每月可支配金額15000以上，大部分持有機車，有汽車，有機車駕照，最常使用機車跟小客車，其他交通工具也有，認知層度(環境、風險、價格都沒有0，新服務有0)，使用目的娛樂>通勤>觀光>洽公

## 解讀命名:
1. 第一群男性較多，收入較高，又具有汽車，因此名為壯年族群
2. 第二群的市場雖然跟第三群很近，但使用人數最少，又沒使用意願，所以命保守族群
3. 第三群的市場可能以學生為主，所以命為:學生上班族族群
4. 第四群的市場以退休或年紀大的人為主，所以命為中老年族群
